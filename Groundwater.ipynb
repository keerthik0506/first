```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("2011.png")
image = cv2.resize(image, (600, 600))

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
blur = cv2.GaussianBlur(image_hsv, (5,5), 0)

lower_blue = np.array([90, 50, 50])
upper_blue = np.array([140, 255, 255])

mask = cv2.inRange(blur, lower_blue, upper_blue)

kernel = np.ones((5,5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

plt.imshow(mask, cmap='gray')
plt.title("Detected Water Area")
plt.show()

water_pixels = np.sum(mask == 255)
total_pixels = mask.size

percentage = (water_pixels / total_pixels) * 100
print("Water Area Percentage:", percentage)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow   #  Import at top
from google.colab import files
uploaded = files.upload()


# Store year-wise results
years = ["2011", "2015", "2019", "2020", "2024"]
water_percentages = []

for year in years:

    # 1 Load Image
    image = cv2.imread(f"{year}.png")

    if image is None:
        print(f"Image {year}.png not found!")
        continue

    # 2 Convert to HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # 3 Blue Color Threshold
    lower_blue = np.array([90, 50, 50])
    upper_blue = np.array([130, 255, 255])

    # 4 Create Mask
    mask = cv2.inRange(hsv, lower_blue, upper_blue)

    # Highlight water body
    highlighted = cv2.bitwise_and(image, image, mask=mask)

    # 5 Count White Pixels
    white_pixels = cv2.countNonZero(mask)
    total_pixels = image.shape[0] * image.shape[1]

    # 6 Calculate Percentage
    water_percentage = (white_pixels / total_pixels) * 100
    water_percentages.append(water_percentage)

    print(f"{year} Water Area: {water_percentage:.2f}%")

    #  Show each year result
    print(f"Highlighted Water - {year}")
    cv2_imshow(highlighted)
    print("\n")

# 8 Compare Results
initial = water_percentages[0]
final = water_percentages[-1]

reduction = ((initial - final) / initial) * 100
print(f"\nWater body reduced by {reduction:.2f}%")

# Risk Message
if reduction > 30:
    print("Groundwater recharge risk: HIGH")
elif reduction > 15:
    print("Groundwater recharge risk: MODERATE")
else:
    print("Groundwater recharge risk: LOW")

#  Plot Graph
plt.figure()
plt.plot(years, water_percentages, marker='o')
plt.xlabel("Year")
plt.ylabel("Water Area (%)")
plt.title("Water Body Reduction Trend")
plt.show()

```
